Thinking1	关联规则中的支持度、置信度和提升度代表的什么，如何计算	

支持度指的是商品组合出现的次数和次数的比值，支持度越高，代表这个商品组合出现的概率越高；置信度(A->B)代表的是一个条件概念，表示购买了A商品之后，购买B商品的概率，AB组合出现的概率除以A出现的概率即为所求；支持度(A->B)代表的是A商品出现对B商品出现的概率提升，置信度(A->B)除以P(A)即为所求。

Thinking2	关联规则与协同过滤的区别

关联规则是对过往交易数据挖中掘出规则，它可以是商品同时被购买的规则，也可以是按时间依次被购买商品的序列模型，它与用户偏好无关，而是基于频繁项集，从而根据每个商品组合的支持度、置信度和提升度来获取规则；协同过滤主要有两种，基于用户的协同过滤，它是通过计算与该用户相似度很高的其他用户，从而将其它用户的内容推荐给该用户，另一种是基于商品的协同过滤，通过对商品的相似度来对商品进行推荐，计算出某个用户喜爱的商品，从而计算该商品相似度很高的其他商品，再将这些商品推荐给该用户。关联规则直接从数据中挖掘出规则，不考虑用户的偏爱，忽略个性化的场景，适用于超市购物、交通规划以及汽车导航；而协同过滤的过程是先协同然后再过滤，找到相似的人或物，再过滤，考虑个性化的场景，例如音乐电影的荐，其中基于用户的推荐用于用户的数量远大于物品的数量，例如微博、新闻等信息类的推荐，而基于物品的推荐适用于物品的数量远大于用户的数量，并且物品数量相对稳定且不频繁更新，例如电子商务等在线网站。

Thinking3 为什么我们需要多种推荐算法

不同的场景使用不同的算法，每种推荐算法都有不适用的场景，这时候就需要其他推荐算法做补充，多种推荐算法协同运作才能达到更好的效果和收益。比如说遇到用户冷启动场景时，这时候协同过滤就无法根据新用户的用户行为赖进行推荐，这时候就可以使用其他推荐算法来弥补其缺陷，比如基于内容的推荐算法，将热门内容推荐给新用户。

Thinking4	关联规则中的最小支持度、最小置信度该如何确定

选择支持度的最终目的就是找出同时被购买的两个商品，可以提高我们的推荐转换率，从而增加收入。那么可以选出支持度最高的前n对,通常是选择总对数的万分之一或者是前20个。置信度就是根据某一个条件，得到一个结论的可信程度、可靠程度。在分析支持度得到的前20对商品中，分别计算双向置信度（购买a同时购买b和购买b同时购买a的置信度都要算），然后筛选中前n个置信度较高的置信度，从中共同选出最合理的一对商品的一个置信度做参考。一般最小支持度的范围在1%-50%，最小置信度的范围在50%-100%。

Thinking5	都有哪些常见的回归分析方法，评价指标是什么


根据不同自变量的数量可以分为一元回归分析和多元回归分析；根据回归方程的表现形式可以分为线性回归分析和非线性回归分析。常见的回归分析方法：线性回归方法、逻辑回归方法、多项式回归方法、决策树、K近邻等。一般情况下，我们希望衡量预测值与真实值的差距。其中最为直观的评价指标包括平均绝对误差(MAE)和均方误差(MSE),因为这也是新型回归模型需要优化的目标。MAE和MSE数值大小本身并没有太大的意义，随着样本的增加，其值的大小也会越来越大。而R-squared既考量了回归值和真实值之间的差距，同时也兼顾了问题本身真实值的变动，同样当数据集越来越大，值也会越来月接近1。而Adjusted R-Square则是抵消了样本数量对R-squared的影响。
